```python
def filter_tweets(df):
    return df.loc[...,['id', 'conversation_id', 'created_at', 'date', 'tweet', 'language', 'hashtags']]

def separate_tweets(df):
    """
    Separates a Twitter DataFrame into simple tweets and conversational tweets.

    Args:
        df (pd.DataFrame): The input DataFrame with 'id' and 'conversation_id' columns.

    Returns:
        tuple: A tuple containing two DataFrames: (simple_tweets, conversational_tweets).
    """
    # Simple tweets are where the tweet id equals the conversation id
    simple_tweets = df[df['id'] == df['conversation_id']].reset_index(drop=True)

    # Conversational tweets are where the tweet id is different from the conversation id
    conversational_tweets = df[df['id'] != df['conversation_id']].reset_index(drop=True)

    return simple_tweets, conversational_tweets
```

```python
simple_tweets, conversational_tweets = separate_tweets(filter_tweets(file))
# simple_tweets.head()
# conversational_tweets.head()
simple_tweets.to_csv("NoConversationTweets.csv")
```
